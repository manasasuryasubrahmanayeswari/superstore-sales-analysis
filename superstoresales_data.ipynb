```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path =r"C:\Users\manasa\OneDrive\Desktop\Superstore_sales_data\SuperStore_Sales_Dataset.csv"
sales_data = pd.read_csv(file_path)
```

### Data Cleaning

```python
# Data Cleaning
# Drop unnecessary columns 'ind1' and 'ind2' since they are entirely null
sales_data_cleaned = sales_data.drop(columns=['ind1', 'ind2'])

# Convert 'Order Date' and 'Ship Date' to datetime
sales_data_cleaned['Order Date'] = pd.to_datetime(sales_data_cleaned['Order Date'], format='%d-%m-%Y')
sales_data_cleaned['Ship Date'] = pd.to_datetime(sales_data_cleaned['Ship Date'], format='%d-%m-%Y')

# Check for missing values in the remaining dataset
missing_values = sales_data_cleaned.isnull().sum()
print("Missing Values:\n", missing_values)

# For simplicity, drop rows with any missing values
sales_data_cleaned = sales_data_cleaned.dropna()

# Verify that the dataset is clean
print(sales_data_cleaned.info())
```

### Data Extraction for Trend Analysis

```python
# Extract year and month from 'Order Date' for monthly trend analysis
sales_data_cleaned['Year-Month'] = sales_data_cleaned['Order Date'].dt.to_period('M')

# Group by 'Year-Month' and calculate total sales and profit
monthly_trends = sales_data_cleaned.groupby('Year-Month').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Convert 'Year-Month' back to datetime for plotting
monthly_trends['Year-Month'] = monthly_trends['Year-Month'].dt.to_timestamp()
```

### Exploratory Data Analysis (EDA)

```python
# Summary statistics of numerical columns
print(sales_data_cleaned.describe())
```

### Trend Analysis

##### Data Analysis - Top Products by Sales and Profit

```python


# Data Analysis - Top Products by Sales and Profit
top_products = sales_data_cleaned.groupby('Product Name').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).sort_values(by='Sales', ascending=False).head(10)

# Display the top 10 products by sales and profit
print("\nTop 10 Products by Sales:\n", top_products)

# Plotting the monthly sales and profit trends
plt.figure(figsize=(12, 6))

# Sales Trend
plt.plot(monthly_trends['Year-Month'], monthly_trends['Sales'], label='Sales', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)

# Profit Trend
plt.twinx()
plt.plot(monthly_trends['Year-Month'], monthly_trends['Profit'], label='Profit', color='orange', marker='o')
plt.ylabel('Profit ($)')

plt.legend(loc='upper left')
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = r"C:\Users\manasa\OneDrive\Desktop\Superstore_sales_data\SuperStore_Sales_Dataset.csv"  # Replace with your file path
sales_data = pd.read_csv(file_path)

# Data Cleaning (same as before)
sales_data_cleaned = sales_data.drop(columns=['ind1', 'ind2'])
sales_data_cleaned['Order Date'] = pd.to_datetime(sales_data_cleaned['Order Date'], format='%d-%m-%Y')

# Data Analysis - Top Products by Sales
top_products = sales_data_cleaned.groupby('Product Name').agg({
    'Sales': 'sum'
}).sort_values(by='Sales', ascending=False).head(10)

# Bar Chart for Top 10 Products by Sales
plt.figure(figsize=(12, 6))
top_products['Sales'].plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.ylabel('Sales ($)')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

##### 2)piechart : Sales Distribution by Category

```python
# Data Analysis - Sales Distribution by Category
category_sales = sales_data_cleaned.groupby('Category').agg({
    'Sales': 'sum'
})

# Pie Chart for Sales Distribution by Category
plt.figure(figsize=(8, 8))
category_sales['Sales'].plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Sales Distribution by Category')
plt.ylabel('')  # Hides the y-label for better appearance
plt.show()
```

#### Advanced Analysis: Heatmap of Sales by Category and Sub-Category

```python
# Pivot table to create a matrix of sales by Category and Sub-Category
sales_pivot = sales_data_cleaned.pivot_table(
    values='Sales',
    index='Sub-Category',
    columns='Category',
    aggfunc='sum'
)

# Heatmap of Sales by Category and Sub-Category
plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, annot=True, fmt=".1f", cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Sales by Category and Sub-Category')
plt.show()
```
